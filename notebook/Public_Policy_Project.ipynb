```python
import pandas as pd

raw_df = pd.read_csv('ALL India pincode.csv')

# Standardize the state names to uppercase
raw_df['statename'] = raw_df['statename'].astype(str).str.upper().str.strip()

# Filter for Maharashtra only
maharashtra_df = raw_df[raw_df['statename'] == 'MAHARASHTRA'].copy()

# Check out our new workspace
print(f"Total Post Offices found in Maharashtra: {len(maharashtra_df)}")
print("\n--- Summary of Columns and Missing Data ---")
maharashtra_df.info()
```

```python
office_dict = {
    'BO': 'Branch Office',
    'PO': 'Post Office',
    'HO': 'Head Office'
}
```

```python

# --- PHASE 2: DATA CLEANING ---

# 1. Check exactly how many missing coordinates we have
missing_coords = maharashtra_df[['latitude', 'longitude']].isnull().sum()
print("Missing Data Before Cleaning:")
print(missing_coords)

# 2. Drop rows that do not have geographic coordinates
# (We need these to plot our map in Power BI later)
clean_mh_df = maharashtra_df.dropna(subset=['latitude', 'longitude']).copy()

# 3. Standardize the Office Type column for better dashboard labels
# BO = Branch Office (usually rural), SO = Sub Office, HO = Head Office
office_dict = {'BO': 'Branch Office', 'SO': 'Sub Office', 'HO': 'Head Office'}
clean_mh_df['office_type_full'] = clean_mh_df['officetype'].map(office_dict).fillna(clean_mh_df['officetype'])

# 4. Verify our cleaned dataset
print("\n--- Cleaning Complete ---")
print(f"Total Post Offices ready for mapping: {len(clean_mh_df)}")
print("\nSample of our new, clean data:")
print(clean_mh_df[['officename', 'district', 'office_type_full', 'latitude', 'longitude']].head())
```

```python
# --- PHASE 3: EXPLORATORY ANALYSIS & EXPORT ---

# 1. Summarize the infrastructure distribution by district
district_counts = clean_mh_df['district'].value_counts().reset_index()
district_counts.columns = ['District', 'Total_Post_Offices']

print("--- Top 10 Districts by Postal Infrastructure ---")
print(district_counts.head(10))

print("\n--- Bottom 5 Districts by Postal Infrastructure ---")
print(district_counts.tail(5))

# 2. Save the cleaned dataset to a new CSV file
final_filename = 'Cleaned_Maharashtra_Postal_Data.csv'
clean_mh_df.to_csv(final_filename, index=False)
print(f"\nSuccessfully saved cleaned data to cloud as: {final_filename}")

# 3. Download the file to your local computer (Colab specific command)
from google.colab import files
files.download(final_filename)
```
